## PROCESS DATA REPORT

The `Process Data Report` is composed of a set of functions that run on a pandas Dataframe.
The functions retrieve a comprenhensive report and a basic set of plots that can lead the process of Exploratory Data Analysis and Feature Engineering.

The functions provided by the `Process Data Report` are listed next:
   - report: **data process report generator**
   - codes: **description of codes regarding with process suggestion**
   - corr_matrix: **correlation matrix plot as heatmap**
   - univar_plot: **histogram and QQ-plot for a single variable**
   - multivar_plot: **pairwise plot showing the relationship of the variables**
   - sta_description: **statistical description of a set of variables**
   - variable_type: **separates the variables in the dataset as Numerical / Categorical**
   

```python
##IMPORTS SECTION
import numpy as np
import pandas as pd
from process_data import data_report as rp
%matplotlib inline
```

### EXAMPLE #02

As a case of use #02 on the `Process Data Report` the _House Price Dataset_ is used

```python
#READ DATASET
df_house = pd.read_csv('./data/House.csv')
df_house.head(3)
```

```python
##BASIC REPORT ON THE DATASET
rp.report(df_house)
```

```python
##CORRELATION MATRIX OF THE DATASET
rp.corr_matrix(df_house)
```

Examining the heatmap we can notice several areas that show **high correlation**. In order to get a better undestanding of these variables we're going to take a look on the 10 most correlated variables with the target variable `SalePrice`

```python
var_corr = df_house.corr()
col = var_corr.nlargest(10, 'SalePrice').index
df_10_corr = df_house[col]
rp.corr_matrix(df_10_corr)
```

It can be noticed that not only the variables are correlated with the target but also some of them are correlated with each other so, this suggest that we can infer the same information with only one of them.
Taking this into account we're going to select the next variables:
- OverallQual
- GrLivArea
- GarageCars
- TotalBsmtSF
- YearBuilt

```python
##PAIRWISE PLOT FOR SELECTED VARIABLES
var = ['SalePrice','OverallQual','GrLivArea','GarageCars','TotalBsmtSF','YearBuilt']
sel_var = df_house[var]
rp.multivar_plot(sel_var)
```

```python
##STATISTICAL DESCRIPTION FOR THE VARIABLES SELECTED
rp.sta_description(df_house,var)
```

```python
##ANALYSING VARIABLES DISTRIBUTION OF SOME OF THE SELECTED VARIABLES
var_dist = [sel_var[var].dropna() for var in ['SalePrice','GrLivArea','TotalBsmtSF','YearBuilt']]
for dist in var_dist:
    rp.univar_plot(dist)
```
